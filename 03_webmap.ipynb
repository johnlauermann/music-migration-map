# Global Jazz Diasporas
### Map migration networks
This script creates geospatial data for Cisco Bradley's Global Jazz Diasporas project at the Pratt Institute Music & Migration Lab. It applies cartography and publishes data to a web map. 

### Set up for replication
This script was designed to be run in ArcGIS Pro 3.5 using Python 3. To replicate, create a new ArcGIS project, add this notebook, and then run as follows. 

```python
import arcpy
import os
```

```python
#set up workspaces and check database
aprx = arcpy.mp.ArcGISProject("Current")
default_gdb = aprx.defaultGeodatabase
default_folder = aprx.homeFolder
arcpy.env.overwriteOutput = True
print("Directory: " + default_folder)
print("Geodatabase: " + default_gdb)
```

### Style the cartography
Setting properties here to ensure consistent visual identity. 

```python
# Reference the relevant map
m = aprx.listMaps("web map")[0]
```

```python
# origin points
lyr = m.listLayers("origins")[0]
sym = lyr.symbology

if hasattr(sym, 'renderer'):
    sym.updateRenderer('GraduatedSymbolsRenderer')
    sym.renderer.classificationField = "Path_Count"
    sym.renderer.breakCount = 5
    
    # Push the changes back to the layer
    lyr.symbology = sym

print("symbolized origins")

lyr = m.listLayers("origins_rankedbypaths")[0]
sym = lyr.symbology

if hasattr(sym, 'renderer'):
    sym.updateRenderer('GraduatedSymbolsRenderer')
    sym.renderer.classificationField = "Path_Count"
    sym.renderer.breakCount = 5
    
    # Push the changes back to the layer
    lyr.symbology = sym

print("symbolized ranked origins")
```

```python
# destination points
lyr = m.listLayers("destinations")[0]
sym = lyr.symbology

if hasattr(sym, 'renderer'):
    sym.updateRenderer('GraduatedSymbolsRenderer')
    sym.renderer.classificationField = "Path_Count"
    sym.renderer.breakCount = 5
    
    # Push the changes back to the layer
    lyr.symbology = sym

print("symbolized origins")

lyr = m.listLayers("destinations_rankedbypaths")[0]
sym = lyr.symbology

if hasattr(sym, 'renderer'):
    sym.updateRenderer('GraduatedSymbolsRenderer')
    sym.renderer.classificationField = "Path_Count"
    sym.renderer.breakCount = 5
    
    # Push the changes back to the layer
    lyr.symbology = sym

print("symbolized ranked origins")
```

```python
# paths
lyr = m.listLayers("paths")[0]
sym = lyr.symbology

# change color
lyr_def = lyr.getDefinition("V3")
lyr_def.renderer.symbol.symbol.symbolLayers[0].color.values = [252, 42, 42, 25]
lyr.setDefinition(lyr_def)

print("symbolized paths")
```

```python
# life paths
lyr = m.listLayers("life_paths")[0]
sym = lyr.symbology

lyr_def = lyr.getDefinition("V3")
lyr_def.renderer.symbol.symbol.symbolLayers[0].color.values = [0, 191, 43, 25]

# Apply the changes back to the layer
lyr.setDefinition(lyr_def)

print("symbolized life paths")
```

### Pop ups
Configure popup text for use in web map. 

```python
# origins

## define variables
lyr = m.listLayers("origins")[0]
sym = lyr.symbology
lyr_def = lyr.getDefinition("V3")

## check for popup
if not hasattr(lyr_def, "popupInfo") or lyr_def.popupInfo is None:
    lyr_def.popupInfo = arcpy.cim.CIMPopup.CIMPopupInfo()

## configure properties
lyr_def.popupInfo.title = "Origin: {Depart_City}, {Depart_State}"  
lyr_def.popupElement_text = "{Path_Count} migration paths originated here."
lyr_def.popupInfo.showPopups = True

## update
lyr.setDefinition(lyr_def)
aprx.save()
```
